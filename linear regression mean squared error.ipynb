```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate data
X = np.linspace(0, 10, 50)       # 50 points between 0 and 10
y = 2 * X + 1 + np.random.randn(50)  # y = 2x + 1 + noise
```

```python
plt.scatter(X, y)
plt.xlabel("X")
plt.ylabel("y")
plt.title("Data Points")
plt.show()
```

```python
def mse(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)
```

```python
# Initialize parameters
w = 0.0
b = 0.0
learning_rate = 0.01
epochs = 1000

n = len(X)

# Gradient Descent loop
for i in range(epochs):
    y_pred = w * X + b
    error = y_pred - y

    # Compute gradients
    dw = (2/n) * np.dot(error, X)
    db = (2/n) * np.sum(error)

    # Update parameters
    w -= learning_rate * dw
    b -= learning_rate * db

    # Optional: print progress every 100 steps
    if i % 100 == 0:
        print(f"Epoch {i}: w={w:.3f}, b={b:.3f}, MSE={mse(y, y_pred):.3f}")
```

```python
y_pred_final = w * X + b

plt.scatter(X, y, label='Data')
plt.plot(X, y_pred_final, color='red', label='Best Fit Line')
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()

print(f"Final Model: y = {w:.2f}x + {b:.2f}")
```

```python
final_mse = mse(y, y_pred_final)
print(f"Final Mean Squared Error: {final_mse:.3f}")
```
