```python
import numpy as np
import pandas as pd
from numpy.linalg import inv, slogdet
from math import log, pi
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

# =======================
# 1. Download & Load Iris Data
# =======================

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
# Iris dataset columns (no header in CSV)
columns = ["sepal_length", "sepal_width", "petal_length", "petal_width", "class"]

df = pd.read_csv(url, header=None, names=columns)

print("Dataset shape:", df.shape)
print(df.head())

# Encode class labels to integers
class_mapping = {label: idx for idx, label in enumerate(df['class'].unique())}
df['class_id'] = df['class'].map(class_mapping)

X = df.iloc[:, :-2].values  # Features (exclude class and class_id)
y = df['class_id'].values

print("Class mapping:", class_mapping)

# Split dataset into train and test sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# =======================
# 2. Gaussian Bayes functions (same as before)
# =======================
def fit_gaussian_bayes(X, y, regularize=1e-6):
    classes = np.unique(y)
    params = {}
    for c in classes:
        Xc = X[y == c]
        mean = Xc.mean(axis=0)
        cov = np.cov(Xc, rowvar=False) + np.eye(Xc.shape[1]) * regularize
        prior = len(Xc) / len(X)
        params[c] = {"mean": mean, "cov": cov, "prior": prior}
    return params

def log_multivariate_normal_pdf(x, mean, cov):
    d = len(mean)
    x = np.atleast_2d(x)
    xc = x - mean
    sign, logdet = slogdet(cov)
    inv_cov = inv(cov)
    quad = np.sum((xc @ inv_cov) * xc, axis=1)
    return -0.5 * (d * log(2 * pi) + logdet + quad)

def predict_gaussian_bayes(X, params):
    classes = list(params.keys())
    log_probs = []
    for c in classes:
        p = params[c]
        log_likelihood = log_multivariate_normal_pdf(X, p["mean"], p["cov"])
        log_prior = np.log(p["prior"])
        log_probs.append(log_likelihood + log_prior)
    log_probs = np.column_stack(log_probs)
    return np.array([classes[i] for i in np.argmax(log_probs, axis=1)])

# Train model
params = fit_gaussian_bayes(X_train, y_train)

# Predict
y_pred = predict_gaussian_bayes(X_test, params)

# =======================
# 3. Evaluate Performance
# =======================
accuracy = np.mean(y_pred == y_test)
print("\nAccuracy:", accuracy)

def confusion_matrix(y_true, y_pred):
    classes = np.unique(np.concatenate([y_true, y_pred]))
    matrix = np.zeros((len(classes), len(classes)), dtype=int)
    for t, p in zip(y_true, y_pred):
        i = np.where(classes == t)[0][0]
        j = np.where(classes == p)[0][0]
        matrix[i, j] += 1
    return classes, matrix

classes_, cm = confusion_matrix(y_test, y_pred)
print("\nConfusion Matrix (rows=True, cols=Pred):\n", cm)

plt.figure(figsize=(5, 4))
plt.imshow(cm, cmap="Blues", interpolation="nearest")
plt.title("Confusion Matrix - Bayesian Decision Theory")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.xticks(ticks=np.arange(len(classes_)), labels=[list(class_mapping.keys())[i] for i in classes_])
plt.yticks(ticks=np.arange(len(classes_)), labels=[list(class_mapping.keys())[i] for i in classes_])
for (i, j), val in np.ndenumerate(cm):
    plt.text(j, i, val, ha="center", va="center")
plt.tight_layout()
plt.show()
```
